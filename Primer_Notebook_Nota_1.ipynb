# Problema de ingeniería

**Calcular el área de un terreno rectangular.**

Este cálculo se usa frecuentemente para:

* estimar **área de cultivo**
* calcular **superficie de construcción**
* dimensionar **parcelas o lotes**

La fórmula es:

[
Área = largo \times ancho
]

---

# 1. Definición del algoritmo

## Entradas

Datos que el usuario debe proporcionar.

* `largo` del terreno (metros)
* `ancho` del terreno (metros)

## Proceso

Operación matemática.

* Multiplicar largo por ancho

[
area = largo * ancho
]

## Salida

Resultado que entrega el programa.

* Área del terreno en **m²**

---

# 2. Pseudocódigo

El pseudocódigo describe la lógica **antes de programar**.

```
INICIO

leer largo
leer ancho

area = largo * ancho

mostrar area

FIN
```

---

# 3. Código en Python

```python
# Programa para calcular el área de un terreno rectangular
# Aplicación básica de ingeniería

# ENTRADAS
largo = float(input("Ingrese el largo del terreno (m): "))
ancho = float(input("Ingrese el ancho del terreno (m): "))

# PROCESO
area = largo * ancho

# SALIDA
print("El área del terreno es:", area, "m²")
```

---

# 4. Ejemplo de ejecución

Resultado:

# 6. Mejora típica en ingeniería

Podemos agregar **unidades y mejor formato**:
